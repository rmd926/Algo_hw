```python
import time
import matplotlib.pyplot as plt

# Method 1: Recursive Fibonacci
def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

# Method 2: Dynamic Programming Fibonacci
def fibonacci_dynamic(n):
    fib = [0] * (n+1)
    fib[1] = 1
    for i in range(2, n+1):
        fib[i] = fib[i-1] + fib[i-2]
    return fib[n]

# Measure execution time for both methods
methods = ['Recursive', 'Dynamic Programming']
execution_times = {method: [] for method in methods}
for i in range(1, 51):
    for method in methods:
        start_time = time.time()
        if method == 'Recursive':
            fibonacci_recursive(i)
        elif method == 'Dynamic Programming':
            fibonacci_dynamic(i)
        execution_times[method].append(time.time() - start_time)

# Plotting the results
plt.figure(figsize=(10, 6))
for method in methods:
    plt.plot(range(1, 51), execution_times[method], label=method)
plt.title('Execution Time of Fibonacci Calculation Methods')
plt.xlabel('N (Fibonacci Number Index)')
plt.ylabel('Execution Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```
