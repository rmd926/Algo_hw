```python
import matplotlib.pyplot as plt

# Recursive solution
def fibonacci_recursive(n, counts):
    if n <= 1:
        return n
    counts[0] += 1
    return fibonacci_recursive(n-1, counts) + fibonacci_recursive(n-2, counts)

# Dynamic programming solution
def fibonacci_dp(n):
    fib = [0] * (n + 1)
    fib[1] = 1
    for i in range(2, n + 1):
        fib[i] = fib[i - 1] + fib[i - 2]
    return fib[n]

# Count number of times F(4) is computed
def count_f4_recursive(n):
    counts = [0]
    fibonacci_recursive(n, counts)
    return counts[0]

def count_f4_dp(n):
    counts = [0] * (n + 1)
    fibonacci_dp(n)
    return counts[4]

# Calculate counts for both methods
n_values = list(range(5, 51))
counts_recursive = [count_f4_recursive(n) for n in n_values]
counts_dp = [count_f4_dp(n) for n in n_values]

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(n_values, counts_recursive, label='Recursive', marker='o')
plt.plot(n_values, counts_dp, label='Dynamic Programming', marker='o')
plt.title('Number of times F(4) is computed')
plt.xlabel('Value of n')
plt.ylabel('Number of times F(4) is computed')
plt.legend()
plt.grid(True)
plt.show()
```
